```
# returns the same vector vec with the i-th and j-th entries swapped
def swap(vec, i, j):
  temp = vec[i]
  vec[i] = vec[j]
  vec[j] = temp
  return vec
```

```
# returns a list containing
def partition(vec):
  n = len(vec)
  if n == 1:
    return [vec, 0]
  else:
    pivot = vec[n - 1]
    i = 0
    j = n - 2
    while i < j:
      while (vec[i] <= pivot) and (i < n - 1):
        i = i + 1
        if i == n - 1:
          return [vec, n - 1]
      while (vec[j] > pivot) and (j > 0):
        j = j - 1
      if i >= j:
        swap(vec, i, n - 1)
        return [vec, i]
      else:
        swap(vec, i, j)
```

```
partition([5, 4, 3, 2, 1])
```
